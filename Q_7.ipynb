```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('us-marriages-divorces-1867-2014.csv')

df.head(), df.columns
```

```python
# Selecting years 1900, 1950, and 2000
selected_years_new = [1900, 1950, 2000]
data_filtered_new = df[df['Year'].isin(selected_years_new)]

plt.figure(figsize=(10, 6))
width_new = 0.35 
index_new = np.arange(len(selected_years_new))

bar1_new = plt.bar(index_new - width_new/2, data_filtered_new['Marriages_per_1000'], width_new, label='Marriages per 1000')
bar2_new = plt.bar(index_new + width_new/2, data_filtered_new['Divorces_per_1000'], width_new, label='Divorces per 1000')


plt.xlabel('Year')
plt.ylabel('Count per 1000 People')
plt.title('Comparison of Marriages and Divorces per Capita in the U.S. for 1900, 1950, and 2000')
plt.xticks(index_new, selected_years_new)
plt.legend()

plt.show()
```
